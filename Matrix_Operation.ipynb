```python
#Multiplication of Matrix:-
import math
def matrix(m,n):
    matrix_output=[]
    for i in range (m):
        row=[]
        for j in range (n):
            inp=int(input(f"enter matrix_output[{i}][{j}]"))
            row.append(inp)
        matrix_output.append(row)
    return matrix_output
def multiplication(A,B):
    output=[]
    for i in range(len(A)):
        row=[]
        for j in range(len(A[0])):
            element=0
            for k in range(len(B)):
                element += A[i][k] *B[k][j]
            row.append(element)
        output.append(row)
    return output
m=int(input("Enter the value of m\n"))
n=int(input("enter the value of n\n"))       

print("enter the matrix A")
A = matrix(m,n)
print(A)
print()

print("enter the matrix B")
B = matrix(m,n)
print(B)
print()

print("Multiplication of Matrix A and B : ")
M= multiplication(A,B)
print(M)
```
